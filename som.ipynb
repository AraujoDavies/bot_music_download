```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep
# from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.service import Service
import os
import json

# Sessão de variaveis
path_to_downloads = "C:\\Users\\MrRobot\\Downloads\\"
path_to_musics = "C:\\Users\\MrRobot\\Music\\"
service = Service(executable_path="./chromedriver.exe")
options = webdriver.ChromeOptions()
user_agent = 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/60.0.3112.50 Safari/537.36' # se tiver "headless" no user_agent o site bloqueia o acesso
options.add_argument('user-agent={0}'.format(user_agent))
# options.add_argument('--headless')
# options.add_argument('--no-sandbox')
# options.add_argument('--mute-audio')
# options.add_argument("user-data-dir=C:\\Users\\MrRobot\\AppData\\Local\\Google\\Chrome\\User Data\\Profile 4") #Path to your chrome profile
c = webdriver.Chrome(service=service, options=options)
```

```python
def valida_tamanho_do_arquivo(mb):
    """ Recebe uma música do diretório"""
    x = os.path.getsize("C:\\Users\\MrRobot\\Downloads\\"+mb)
    if x < 312000:
        return False
    return True

def mover_msc_to_paste(webm = False):
    """parametro webm utilizado para downloads com extensão .webm"""
    try:
        mb = os.listdir("C:\\Users\\MrRobot\\Downloads\\")
        mb.remove('desktop.ini')
        x = 0
        while valida_tamanho_do_arquivo(mb[0]) == False: # ainda não baixou
            sleep(2)
            print('download em andamento')
            x += 1
            if x > 10:
                print('download não concluído ou não iniciado')
                break
        if valida_tamanho_do_arquivo(mb[0]):
            if webm:
                os.replace(f"C:\\Users\\MrRobot\\Downloads\\{mb[0]}", f"C:\\Users\\MrRobot\\Music\\{mb[0].replace('.webm', '.mp3')}")
            else:
                os.replace(f"C:\\Users\\MrRobot\\Downloads\\{mb[0]}", f"C:\\Users\\MrRobot\\Music\\{mb[0]}")
    except:
        print('download não concluído ou não iniciado')
    else:
        mb2 = os.listdir("C:\\Users\\MrRobot\\Downloads\\")
        mb2.remove('desktop.ini')
        if len(mb2) == 0:
            print(f'arquivo {mb[0]} movido')
```

```python
def download_wait(path_to_downloads):
    seconds = 0
    download_wait = True
    path_to_downloads.remove('desktop.ini')
    while download_wait and seconds < 20:
        sleep(1)
        download_wait = False
        for fname in os.listdir(path_to_downloads):
            print(fname)
            if valida_tamanho_do_arquivo(path_to_downloads+fname):
                download_wait = True
        seconds += 1
    return seconds

download_wait(path_to_downloads)
```

```python
def pegando_urls_no_ytb(link):
    try:
        c.get(link)

        videos = c.find_elements(By.XPATH, "//a[@id = 'video-title']")
    
        urls = []
        for hrefs in videos:
            dict = {}
            dict['title'] = hrefs.get_attribute("innerHTML").replace('\n          ', '').replace('\n        ', '')
            dict['url'] = hrefs.get_attribute("href")
            urls.append(dict)

        return urls
    except:
        return "ERRO ao pegar URLS da playlist no Youtube."
```

```python
def automacao_de_download_yt1s(playlist):
  array_erros = []
  for msc in playlist:
    try:
      # print(url)
      site_mp3 = 'https://yt1s.com/pt374/youtube-to-mp3'
      c.get(site_mp3)
      
      search_bar = c.find_element(By.XPATH, '//input[@id = "s_input"]')
      search_bar.send_keys(msc['url'])

      btn_convert = c.find_element(By.XPATH, '//button[@id = "btn-convert"]')
      btn_convert.click()

      c.implicitly_wait(10)
      get_link = c.find_element(By.XPATH, '//button[@id = "btn-action"]')
      get_link.click()

      c.implicitly_wait(10)
      baixar = c.find_element(By.XPATH, '//a[@id = "asuccess"]')
      baixar.click()

      sleep(10)

      mb = os.listdir("C:\\Users\\MrRobot\\Downloads\\")
      mb.remove('desktop.ini')

      if mb == 0: # download deu bosta
        array_erros.append(msc['url'])
      else:
        os.replace(f"C:\\Users\\MrRobot\\Downloads\\{mb[0]}", f"C:\\Users\\MrRobot\\Music\\{mb[0]}")
    except:
      array_erros.append(msc['url'])
  
  c.quit()
  return array_erros
```

```python
def automacaoo_de_download_flv(array_de_urls):
    site = "https://flvto.pro/"
    c.get(site)
    array_erros = []
    for msc in array_de_urls:
        try:
            search_bar = c.find_element(By.ID, "input")
            search_bar.send_keys(msc)
                
            convert_btn = c.find_element(By.CLASS_NAME, "conver-btn")
            convert_btn.click()

            c.implicitly_wait(10)
            download_btn = c.find_element(By.CLASS_NAME, "download-btns")
            download_btn.click()

            mb = os.listdir(path_to_downloads)
            mb.remove('desktop.ini')

            if mb == 0: # download deu bosta
                array_erros.append(msc)
            else:
                os.replace(f"{path_to_downloads}{mb[0]}", f"{path_to_musics}{mb[0].replace('.webm', '.mp3')}")
        except:
            array_erros.append(msc)

        return array_erros
```
