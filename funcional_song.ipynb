```python
def pegando_urls_no_ytb(link):
    try:
        c.get(link)

        videos = c.find_elements(By.XPATH, "//a[@id = 'video-title']")
    
        urls = []
        for hrefs in videos:
            dict = {}
            dict['title'] = hrefs.get_attribute("innerHTML").replace('\n          ', '').replace('\n        ', '')
            dict['url'] = hrefs.get_attribute("href")
            urls.append(dict)

        return urls
    except:
        return "ERRO ao pegar URLS da playlist no Youtube."
```

```python
def automacao_de_download_yt1s(playlist):
#  array_erros = []
  for msc in playlist:
    try:
      mb = os.listdir(path_to_downloads)
      mb.remove('desktop.ini')
      qt_mscs_antes_download = len(mb)

      print(f'iniciando download da {msc["title"]}')
      site_mp3 = 'https://yt1s.com/pt374/youtube-to-mp3'
      c.get(site_mp3)

      search_bar = c.find_element(By.XPATH, '//input[@id = "s_input"]')
      search_bar.send_keys(msc['url'])

      btn_convert = c.find_element(By.XPATH, '//button[@id = "btn-convert"]')
      btn_convert.click()

      c.implicitly_wait(10)
      get_link = c.find_element(By.XPATH, '//button[@id = "btn-action"]')
      get_link.click()

      c.implicitly_wait(20)
      baixar = c.find_element(By.XPATH, '//a[@id = "asuccess"]')
      baixar.click()

      sleep(15)

      mb = os.listdir(path_to_downloads)
      mb.remove('desktop.ini')
      qt_mscs_apos_download = len(mb)

      if qt_mscs_apos_download != qt_mscs_antes_download:
          print(f'Sucesso no download da {msc["title"]}')
          playlist.remove(msc)
          os.replace(f"{path_to_downloads}{mb[0]}", f"{path_to_musics}{mb[0]}")        
          
      # if len(mb) == 0: # download deu bosta
      #   print(f'Falha no download da {msc["title"]}')
      #   array_erros.append(msc)
      # elif len(mb) == 1:
      #   print(f'Sucesso no download da {msc["title"]}')
      #   os.replace(f"{path_to_downloads}{mb[0]}", f"{path_to_musics}{mb[0]}")
    except:
      print(f'Falha no download da {msc["title"]}')
      # array_erros.append(msc)
  
  return playlist
```

```python
def automacaoo_de_download_flv(playlist):
    site = "https://flvto.pro/"
    array_erros = []
    array_sucesso = []
    for msc in playlist:
        try:
            c.get(site)
            mb = os.listdir(path_to_downloads)
            mb.remove('desktop.ini')
            qt_mscs_antes_download = len(mb)

            print(f'iniciando download da {msc["title"]}')
            search_bar = c.find_element(By.ID, "input")
            search_bar.send_keys(msc['url'])
                
            convert_btn = c.find_element(By.CLASS_NAME, "conver-btn")
            convert_btn.click()

            c.implicitly_wait(20)
            download_btn = c.find_element(By.CLASS_NAME, "download-btns")
            download_btn.click()

            sleep(12)

            mb = os.listdir(path_to_downloads)
            mb.remove('desktop.ini')
            qt_mscs_apos_download = len(mb)

            if qt_mscs_apos_download != qt_mscs_antes_download:
                print(f'Sucesso no download da {msc["title"]}')
                os.replace(f"{path_to_downloads}{mb[0]}", f"{path_to_musics}{mb[0].replace('.webm', '.mp3')}")
                array_sucesso.append(msc)
            else:
                array_erros.append(msc)
        
            #CTRL + W
                    
        except:
            print(f'Falha no download da {msc["title"]}')
            array_erros.append(msc)

    
    return array_sucesso, array_erros
```
